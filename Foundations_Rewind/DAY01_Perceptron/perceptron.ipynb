🧠 Perceptron from Scratch and compare with Sklearn

```python
import numpy as np
from sklearn.linear_model import Perceptron

# -----------------------------
# 1. Dataset (AND Gate)
# -----------------------------
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# For scratch model
y = np.array([-1, -1, -1, +1])

# For sklearn (0/1 format)
y_sklearn = np.array([0, 0, 0, 1])


# -----------------------------
# 2. Perceptron From Scratch
# -----------------------------
class PerceptronScratch:
    def __init__(self, lr=0.1, epochs=10):
        self.lr = lr
        self.epochs = epochs

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for epoch in range(self.epochs):
            errors = 0
            for xi, target in zip(X, y):
                z = np.dot(xi, self.weights) + self.bias
                y_pred = 1 if z >= 0 else -1

                if y_pred != target:
                    self.weights += self.lr * target * xi
                    self.bias += self.lr * target
                    errors += 1

            print(f"Epoch {epoch+1}: Misclassified = {errors}")
            if errors == 0:
                print("✅ Converged")
                break

    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        return np.where(z >= 0, 1, -1)


# -----------------------------
# 3. Train Scratch Model
# -----------------------------
scratch_model = PerceptronScratch(lr=0.1, epochs=10)
scratch_model.fit(X, y)

your_preds = scratch_model.predict(X)

print("\n🔹 Scratch Model")
print("Weights:", scratch_model.weights)
print("Bias:", scratch_model.bias)
print("Predictions:", your_preds)


# -----------------------------
# 4. Train sklearn Model
# -----------------------------
sk_model = Perceptron(eta0=0.1, max_iter=10, tol=None)
sk_model.fit(X, y_sklearn)

sk_preds = sk_model.predict(X)

# Convert sklearn output (0 → -1)
sk_preds_converted = np.where(sk_preds == 0, -1, 1)

print("\n🔹 Sklearn Model")
print("Weights:", sk_model.coef_)
print("Bias:", sk_model.intercept_)
print("Predictions:", sk_preds_converted)


# -----------------------------
# 5. Final Comparison
# -----------------------------
print("\n🔍 Comparison")
print("Scratch:", your_preds)
print("Sklearn:", sk_preds_converted)
```
