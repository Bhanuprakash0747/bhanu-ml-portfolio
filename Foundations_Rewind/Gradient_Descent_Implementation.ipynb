# Gradient Descent Implementation

## Problem Statement
Predict y from X using linear regression

## Model
y = θ0 + θ1x

## Loss Function
Mean Squared Error

```python
import numpy as np

# Simple dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([3, 4, 2, 4, 5], dtype=float)

m = len(X)  # number of samples

theta0 = 0   # intercept
theta1 = 0   # slope
alpha = 0.01 # learning rate
iterations = 2500
losses = []

for i in range(iterations):
    y_pred = theta0 + theta1 * X   

    loss = np.mean((y_pred - y)**2) #measure error loss
    losses.append(loss) #store loss
    
    d_theta0 = (1/m) * np.sum(y_pred - y)
    d_theta1 = (1/m) * np.sum((y_pred - y) * X)
    
    theta0 -= alpha * d_theta0
    theta1 -= alpha * d_theta1

print("Final θ0:", theta0)
print("Final θ1:", theta1)

y_pred_final = theta0 + theta1 * X
mse = np.mean((y_pred_final - y)**2)
print("MSE:", mse)
```
