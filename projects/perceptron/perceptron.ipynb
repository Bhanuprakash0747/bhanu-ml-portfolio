```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([-1,-1,-1,+1])
```

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=1.0, epochs=10):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None   # initialize later
        self.bias = 0.0       # start with bias = 0

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0.0

        for epoch in range(self.epochs):
            errors = 0
            for xi, target in zip(X, y):
                activation = np.dot(xi, self.weights) + self.bias
                y_pred = 1 if activation >= 0 else -1

                if y_pred != target:
                    self.weights += self.lr * target * xi
                    self.bias += self.lr * target
                    errors += 1

            print(f"Epoch {epoch+1}/{self.epochs}, Misclassified: {errors}")
            if errors == 0:
                print("✅ Model has converged (no errors).")
                break

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return np.where(linear_output >= 0, 1, -1)
```

```python
model = Perceptron(learning_rate=0.1, epochs=10)
model.fit(X, y)

print("\nFinal weights:", model.weights)
print("Final bias:", model.bias)
```

```python
import numpy as np
from sklearn.linear_model import Perceptron

# AND gate data
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([0, 0, 0, 1])  # sklearn prefers {0,1} instead of {-1,1}

model = Perceptron(eta0=0.1, max_iter=10, tol=None)
model.fit(X, y)

print("Weights:", model.coef_)
print("Bias:", model.intercept_)

preds = model.predict(X)
print("Predictions:", preds)
```
