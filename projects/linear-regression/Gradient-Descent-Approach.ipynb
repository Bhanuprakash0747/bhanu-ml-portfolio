```python
import numpy as np

# Simple dataset
X = np.array([1, 2, 3, 4, 5], dtype=float) #we are using float because the O/P is expected in Float
y = np.array([3, 4, 2, 4, 5], dtype=float)

m = len(X)  # number of samples
```

```python
theta0 = 0   # intercept
theta1 = 0   # slope
alpha = 0.01 # learning rate
iterations = 2500
```

```python
for i in range(iterations): #iterate 2500 times
    y_pred = theta0 + theta1 * X   # predictions y^i = θ0 + θ1.xi
    #we call it as cost function
    d_theta0 = (1/m) * np.sum(y_pred - y) #mathematical function for optimization
    d_theta1 = (1/m) * np.sum((y_pred - y) * X)
    #parameter update step 
    theta0 -= alpha * d_theta0 # if gradient is +ve subtract and move left - cost decreases
    theta1 -= alpha * d_theta1 # if gradient is -ve addition and move right - cost decreases

print("Final θ0:", theta0)
print("Final θ1:", theta1)
```

```python
y_pred_final = theta0 + theta1 * X
mse = np.mean((y_pred_final - y)**2) #mean squared error 
print("MSE:", mse) #It tells average error per data point (Mistake size)
```
